```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

# Boostrapping

```python
N = 100
dataset = range(1, N+1)
```

```python
np.random.choice(dataset, size=1)
```

```python
sample = np.random.choice(dataset, size=N)
print(f'Sample size:   {len(sample)}')
print(f'Kind of sample: {len(set(sample))}')
```

```python
boostrap_dataset_size = [len(set(np.random.choice(dataset, size=N))) for i in range(N)]
plt.hist(boostrap_dataset_size, bins=100)
plt.show()
```

```python
f = lambda N: (1-1/N)**N
```

```python
N = range(1, 101)
fN = list(map(f, N))
plt.plot(N, fN)
plt.plot(N, np.exp(-1)*np.ones(np.shape(N)))
plt.show()
```

# Bagging

```python
dataset = [-3, -2, -1, 3, 4, 5, 7, 9]
N = len(dataset)
mu = np.mean(dataset)
plt.hist(dataset, bins=50)
plt.show()
```

```python
print(f'mean: {np.mean(dataset)}')
```

```python
boostrap_dataset = np.random.choice(dataset, size=N)
print(boostrap_dataset)
plt.hist(boostrap_dataset, bins=50)
plt.show()
```

```python
boostrap_dataset_mean = [np.mean(np.random.choice(dataset, size=N)) for i in range(1000)]
print(mu)
plt.hist(boostrap_dataset_mean, bins=50)
plt.show()
```
